```python
# ============================================================
# Part 1: Data Acquisition, Cleaning and Exploratory Analysis
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")

# ------------------------------------------------------------
# Task 1: Load Dataset
# ------------------------------------------------------------

df = pd.read_csv("housing.csv")

print("First Five Rows")
print(df.head())

print("\nData Types")
print(df.dtypes)

print("\nShape")
print(df.shape)

# ------------------------------------------------------------
# Task 2: Missing Value Analysis
# ------------------------------------------------------------

null_count = df.isnull().sum()

null_percent = (df.isnull().sum() / df.shape[0]) * 100

null_table = pd.DataFrame({
    "Null Count": null_count,
    "Null Percentage": null_percent
})

print("\nMissing Value Table")
print(null_table)

print("\nColumns with >20% Nulls")
print(null_table[null_table["Null Percentage"] > 20])

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    if null_percent[col] < 20:
        df[col].fillna(df[col].median(), inplace=True)

print("\nRemaining Missing Values")
print(df.isnull().sum())

# ------------------------------------------------------------
# Task 3: Duplicate Removal
# ------------------------------------------------------------

duplicates = df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

before_rows = df.shape[0]

before_null = (df.isnull().sum() / df.shape[0]) * 100

df = df.drop_duplicates()

after_rows = df.shape[0]

removed = before_rows - after_rows

after_null = (df.isnull().sum() / df.shape[0]) * 100

print("Rows Removed:", removed)

comparison = pd.DataFrame({
    "Before": before_null,
    "After": after_null
})

print(comparison)

# ------------------------------------------------------------
# Task 4: Data Type Correction
# ------------------------------------------------------------

memory_before = df.memory_usage(deep=True).sum()

object_cols = df.select_dtypes(include="object").columns

for col in object_cols:

    try:
        converted = pd.to_numeric(df[col], errors="coerce")

        if converted.notna().sum() > 0:
            df[col] = converted
            break

    except:
        pass

if "ocean_proximity" in df.columns:
    df["ocean_proximity"] = df["ocean_proximity"].astype("category")

memory_after = df.memory_usage(deep=True).sum()

print("\nMemory Before:", memory_before)
print("Memory After :", memory_after)

print(df.dtypes)

# ------------------------------------------------------------
# Task 5: Descriptive Statistics & Skewness
# ------------------------------------------------------------

print(df.describe())

skewness = {}

for col in numeric_cols:
    skewness[col] = df[col].skew()

skew_df = pd.DataFrame(skewness.items(),
                       columns=["Column", "Skewness"])

print(skew_df)

highest_skew = skew_df.iloc[
    skew_df["Skewness"].abs().idxmax()
]

print("\nHighest Absolute Skew")
print(highest_skew)

# ------------------------------------------------------------
# Task 6: IQR Outlier Detection
# ------------------------------------------------------------

numeric = df.select_dtypes(include=np.number).columns[:2]

for col in numeric:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print("\nColumn:", col)
    print("Q1:", Q1)
    print("Q3:", Q3)
    print("IQR:", IQR)
    print("Lower:", lower)
    print("Upper:", upper)
    print("Outliers:", len(outliers))

# ------------------------------------------------------------
# Task 7A: Line Plot
# ------------------------------------------------------------

line_col = numeric_cols[0]

plt.figure(figsize=(10,5))
plt.plot(df[line_col])
plt.title(f"Line Plot of {line_col}")
plt.xlabel("Index")
plt.ylabel(line_col)
plt.show()

# ------------------------------------------------------------
# Task 7B: Bar Chart
# ------------------------------------------------------------

if "ocean_proximity" in df.columns:

    bar = df.groupby("ocean_proximity")[numeric_cols[0]].mean()

    plt.figure(figsize=(8,5))
    plt.bar(bar.index.astype(str), bar.values)
    plt.title("Mean by Ocean Proximity")
    plt.xlabel("Ocean Proximity")
    plt.ylabel("Mean")
    plt.xticks(rotation=20)
    plt.show()

# ------------------------------------------------------------
# Task 7C: Histogram
# ------------------------------------------------------------

most_skewed = highest_skew["Column"]

plt.figure(figsize=(8,5))
sns.histplot(df[most_skewed], bins=20)
plt.title(f"Histogram of {most_skewed}")
plt.show()

# ------------------------------------------------------------
# Task 7D: Scatter Plot
# ------------------------------------------------------------

plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x=numeric_cols[0],
    y=numeric_cols[1]
)

plt.title("Scatter Plot")
plt.show()

# ------------------------------------------------------------
# Task 7E: Box Plot
# ------------------------------------------------------------

if "ocean_proximity" in df.columns:

    plt.figure(figsize=(10,5))

    sns.boxplot(
        data=df,
        x="ocean_proximity",
        y=numeric_cols[0]
    )

    plt.title("Box Plot")
    plt.xticks(rotation=25)
    plt.show()

# ------------------------------------------------------------
# Task 8: Pearson Correlation
# ------------------------------------------------------------

pearson = df.corr(numeric_only=True)

print("\nPearson Correlation")
print(pearson)

plt.figure(figsize=(12,8))

sns.heatmap(
    pearson,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()

corr_pairs = pearson.abs().unstack()

corr_pairs = corr_pairs[corr_pairs < 1]

highest_pair = corr_pairs.idxmax()

print("\nHighest Correlated Pair")
print(highest_pair)
print(corr_pairs.max())

# ------------------------------------------------------------
# Task 9A: Mean vs Median
# ------------------------------------------------------------

top2 = skew_df.reindex(
    skew_df["Skewness"].abs().sort_values(
        ascending=False
    ).index
).head(2)

print("\nTop Two Skewed Columns")

print(top2)

for col in top2["Column"]:

    print("\nColumn:", col)

    print("Mean :", df[col].mean())
    print("Median:", df[col].median())

    df[col].fillna(df[col].median(), inplace=True)

print("\nNull Values After Imputation")
print(df[top2["Column"]].isnull().sum())

# ------------------------------------------------------------
# Task 9B: Spearman Correlation
# ------------------------------------------------------------

spearman = df.corr(
    method="spearman",
    numeric_only=True
)

print("\nSpearman Correlation")
print(spearman)

difference = (spearman - pearson).abs()

print("\nDifference Matrix")
print(difference)

pairs = []

cols = difference.columns

for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        pairs.append([
            cols[i],
            cols[j],
            pearson.iloc[i,j],
            spearman.iloc[i,j],
            difference.iloc[i,j]
        ])

difference_table = pd.DataFrame(
    pairs,
    columns=[
        "Column1",
        "Column2",
        "Pearson",
        "Spearman",
        "Difference"
    ]
)

difference_table = difference_table.sort_values(
    by="Difference",
    ascending=False
)

print("\nTop Three Differences")
print(difference_table.head(3))

# ------------------------------------------------------------
# Task 9C: Group Aggregation
# ------------------------------------------------------------

if "ocean_proximity" in df.columns:

    grouped = df.groupby("ocean_proximity")[numeric_cols[0]].agg(
        ["mean","std","count"]
    )

    print(grouped)

    highest_mean = grouped["mean"].idxmax()
    highest_std = grouped["std"].idxmax()

    ratio = grouped["mean"].max()/grouped["mean"].min()

    print("\nHighest Mean Group:", highest_mean)
    print("Highest Std Group :", highest_std)
    print("Mean Ratio:", ratio)

# ------------------------------------------------------------
# Save Clean Dataset
# ------------------------------------------------------------

df.to_csv("cleaned_data.csv", index=False)

print("\ncleaned_data.csv has been created successfully.")
```
